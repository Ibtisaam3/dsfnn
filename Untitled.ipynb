```python
def make_spiral(n_points=1000, noise=0.2):
    n = np.sqrt(np.random.rand(n_points)) * 780 * (2*np.pi)/360
    d1x = -np.cos(n)*n + np.random.rand(n_points) * noise
    d1y = np.sin(n)*n + np.random.rand(n_points) * noise
    X = np.vstack((np.append(d1x, d1x*-1), np.append(d1y, d1y*-1))).T
    y = np.hstack((np.zeros(n_points), np.ones(n_points)))
    return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32).reshape(-1,1)

X, y = make_spiral(1000)
plt.figure(figsize=(5,5))
plt.scatter(X[:,0], X[:,1], c=y.squeeze(), cmap="coolwarm", s=8)
plt.title("Two-spiral dataset")
plt.show()
```

```python
class Net(nn.Module):
    def __init__(self, d_in=2, d_h=16, d_out=1):
        super().__init__()
        self.fc1 = nn.Linear(d_in, d_h)
        self.fc2 = nn.Linear(d_h, d_out)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

model = Net()
print(model)
```

```python
def train_and_track(model, X, y, lr=1e-2, epochs=30, batch_size=128):
    dataset = TensorDataset(X, y)
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.BCELoss()
    
    weight_history = []
    loss_history = []

    for epoch in range(epochs):
        total_loss = 0.0
        for xb, yb in loader:
            pred = model(xb)
            loss = loss_fn(pred, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            total_loss += loss.item()
        
        loss_avg = total_loss / len(loader)
        loss_history.append(loss_avg)

        with torch.no_grad():
            weights = torch.cat([p.flatten() for p in model.parameters()]).detach().cpu()
            weight_history.append(weights)

        print(f"Epoch {epoch+1:02d} | Loss: {loss_avg:.4f}")
    
    return torch.stack(weight_history), loss_history

weight_history, loss_history = train_and_track(model, X, y, lr=1e-2, epochs=50)
```

```python
# Tangent dynamics utilities (uses PyTorch autograd)
import torch, numpy as np

def flatten_params(params):
    return torch.cat([p.contiguous().view(-1) for p in params])

def unflatten_like(vec, like_params):
    out, idx = [], 0
    for p in like_params:
        n = p.numel()
        out.append(vec[idx:idx+n].view_as(p))
        idx += n
    return out

def hvp(loss, params, v_like):
    """Pearlmutter Hessian-vector product: H * v, v is list shaped like params."""
    grad = torch.autograd.grad(loss, params, create_graph=True, retain_graph=True)
    flat_grad = flatten_params(grad)
    v_flat = flatten_params(v_like)
    Hv = torch.autograd.grad(flat_grad, params, grad_outputs=v_flat, retain_graph=True, allow_unused=True)
    Hv = [torch.zeros_like(p) if h is None else h for h in Hv]
    return Hv

class LyapTracker:
    """
    Benettin method with periodic QR:
    propagate k tangent directions through J_t = (I - lr*H_t) for plain SGD.
    """
    def __init__(self, params, k=3, device="cpu", eps=1e-9):
        self.device = device
        self.params = list(params)
        self.dim = sum(p.numel() for p in self.params)
        self.k = min(k, self.dim)
        Q = torch.randn(self.dim, self.k, device=device)
        Q, _ = torch.linalg.qr(Q)
        self.Q = Q
        self.sum_logs = torch.zeros(self.k, device=device)
        self.steps = 0
        self.eps = eps

    @torch.no_grad()
    def _flat_to_like(self, flat):
        return unflatten_like(flat, self.params)

    def step(self, batch_loss, lr):
        # propagate each tangent direction through J = I - lr*H
        new_cols = []
        for j in range(self.k):
            qj = self.Q[:, j]
            v_like = self._flat_to_like(qj)
            Hv_like = hvp(batch_loss, self.params, v_like)
            Hv_flat = flatten_params(Hv_like)
            new_cols.append(qj - lr * Hv_flat)
        M = torch.stack(new_cols, dim=1)   # (dim, k)

        # QR and accumulate log-stretches
        Q, R = torch.linalg.qr(M, mode='reduced')
        diag = torch.diag(R).abs().clamp_min(self.eps)
        self.sum_logs += torch.log(diag)
        self.Q = Q
        self.steps += 1

    def exponents(self):
        if self.steps == 0:
            return torch.zeros(self.k, device=self.device)
        return self.sum_logs / self.steps

    def hks_proxy(self):
        lam = self.exponents()
        return lam.clamp(min=0).sum()
```

```python
from torch.utils.data import DataLoader, TensorDataset
import torch.nn as nn
import matplotlib.pyplot as plt

def train_with_true_lyap(model, X, y, lr=1e-2, epochs=30, batch_size=128, k_lyap=3, micro=64, device="cpu"):
    model = model.to(device)
    ds = TensorDataset(X, y)
    dl = DataLoader(ds, batch_size=batch_size, shuffle=True)
    crit = nn.BCELoss()
    opt = torch.optim.SGD(model.parameters(), lr=lr)  # SGD so J = I - lr*H

    lyap = LyapTracker(model.parameters(), k=k_lyap, device=device)
    logs = {"train_loss": [], "lambda": [], "hks": [], "grad_norm": []}

    for ep in range(1, epochs+1):
        model.train()
        total, gnorm_acc, n_batches = 0.0, 0.0, 0
        for xb, yb in dl:
            xb, yb = xb.to(device), yb.to(device)
            # forward/backward/update 
            opt.zero_grad(set_to_none=True)
            pred = model(xb)
            loss = crit(pred, yb)
            loss.backward()
            # grad norm (dissipation proxy)
            grad_flat = flatten_params([p.grad for p in model.parameters() if p.grad is not None])
            gnorm_acc += torch.norm(grad_flat).item()
            opt.step()
            total += loss.item(); n_batches += 1

            # tangent step (micro-batch)
            xb_m, yb_m = xb[:min(micro, xb.size(0))], yb[:min(micro, yb.size(0))]
            pred_m = model(xb_m)
            loss_m = crit(pred_m, yb_m)
            lyap.step(loss_m, lr=lr)

        train_loss = total / max(1, n_batches)
        lam = lyap.exponents().detach().cpu().numpy()
        hks = lyap.hks_proxy().item()
        logs["train_loss"].append(train_loss)
        logs["lambda"].append(lam.copy())
        logs["hks"].append(hks)
        logs["grad_norm"].append(gnorm_acc / max(1, n_batches))
        print(f"[{ep:02d}] loss={train_loss:.4f}  λ1={lam[0]:+.4f}  hKS≈{hks:.4f}  ⟨||∇||⟩={logs['grad_norm'][-1]:.3f}")
    return model, logs

model_true, logs_true = train_with_true_lyap(Net(), X, y, lr=1e-2, epochs=30, batch_size=128, k_lyap=3)
```

```python
def run_lr(lr, epochs=20, k=1):
    _, logs = train_with_true_lyap(Net(), X, y, lr=lr, epochs=epochs, k_lyap=k)
    return logs

lrs = [5e-3, 1e-2, 2e-2, 5e-2]
final_lam1, final_hks, final_loss = [], [], []

for lr in lrs:
    logs = run_lr(lr, epochs=15, k=1)
    final_lam1.append(logs["lambda"][-1][0])
    final_hks.append(logs["hks"][-1])
    final_loss.append(logs["train_loss"][-1])

plt.figure(figsize=(7,4))
plt.plot(lrs, final_lam1, marker='o'); plt.xscale('log')
plt.xlabel("Learning rate"); plt.ylabel("Final λ1 (per epoch)"); plt.title("Stability vs LR")
plt.grid(True); plt.show()

plt.figure(figsize=(7,4))
plt.plot(lrs, final_loss, marker='o'); plt.xscale('log')
plt.xlabel("Learning rate"); plt.ylabel("Final loss"); plt.title("Loss vs LR")
plt.grid(True); plt.show()

list(zip(lrs, np.round(final_lam1,4), np.round(final_loss,4)))
```

```python
def run_hidden(h, lr=1e-2, epochs=15, k=1):
    m = Net(d_h=h)
    _, logs = train_with_true_lyap(m, X, y, lr=lr, epochs=epochs, k_lyap=k)
    return logs["lambda"][-1][0], logs["train_loss"][-1]

hs = [8, 16, 32, 64]
lam_last, loss_last = [], []
for h in hs:
    lam, los = run_hidden(h, lr=1e-2, epochs=15, k=1)
    lam_last.append(lam); loss_last.append(los)

plt.figure(figsize=(7,4))
plt.plot(hs, lam_last, marker='o')
plt.xlabel("Hidden size"); plt.ylabel("Final λ1"); plt.title("Stability vs width")
plt.grid(True); plt.show()

plt.figure(figsize=(7,4))
plt.plot(hs, loss_last, marker='o')
plt.xlabel("Hidden size"); plt.ylabel("Final loss"); plt.title("Loss vs width")
plt.grid(True); plt.show()

list(zip(hs, np.round(lam_last,4), np.round(loss_last,4)))
```

```python
import torch, numpy as np
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

def train_with_true_lyap_temp(
    model, X, y,
    lr=1e-2, epochs=20, batch_size=128,
    k_lyap=3, micro=64, device="cpu",
    T=0.0,           # temperature (noise strength). 0.0 = no noise
    seed=None,       # set for reproducibility of shuffling/noise
    noise_scale="absgrad"  # "absgrad" or "const"
):
    """
    Adds Gaussian noise to gradients before the optimizer step:
        g_tilde = g + sqrt(2*T) * ξ
    Where ξ ~ N(0, I) scaled either by average |grad| ("absgrad") or a constant ("const").
    """
    if seed is not None:
        torch.manual_seed(seed)
        np.random.seed(seed)

    model = model.to(device)
    ds = TensorDataset(X, y)
    dl = DataLoader(ds, batch_size=batch_size, shuffle=True)
    crit = nn.BCELoss()
    opt  = torch.optim.SGD(model.parameters(), lr=lr)  # keep SGD so J = I - lr*H

    # reuse LyapTracker from earlier
    lyap = LyapTracker(model.parameters(), k=k_lyap, device=device)

    logs = {"train_loss": [], "lambda": [], "hks": [], "grad_norm": [], "T": T}

    for ep in range(1, epochs+1):
        model.train()
        total, gnorm_acc, n_batches = 0.0, 0.0, 0
        for xb, yb in dl:
            xb, yb = xb.to(device), yb.to(device)

            #  forward/backward 
            opt.zero_grad(set_to_none=True)
            pred = model(xb)
            loss = crit(pred, yb)
            loss.backward()

            # measure grad norm (dissipation proxy) 
            grads = [p.grad for p in model.parameters() if p.grad is not None]
            flatg = flatten_params(grads)
            gnorm = torch.norm(flatg).item()
            gnorm_acc += gnorm

            #Langevin-style gradient noise (temperature T) ----
            if T > 0:
                if noise_scale == "absgrad":
                    # scale noise by average absolute grad magnitude (keeps units reasonable)
                    scale = flatg.abs().mean().item() + 1e-12
                else:
                    scale = 1.0
                std = (2.0 * T) ** 0.5 * scale
                with torch.no_grad():
                    for p in model.parameters():
                        if p.grad is not None:
                            p.grad.add_(torch.randn_like(p.grad) * std)

            opt.step()
            total += loss.item(); n_batches += 1

            #  tangent step (micro-batch)
            xb_m, yb_m = xb[:min(micro, xb.size(0))], yb[:min(micro, yb.size(0))]
            pred_m = model(xb_m)
            loss_m = crit(pred_m, yb_m)
            lyap.step(loss_m, lr=lr)

        train_loss = total / max(1, n_batches)
        lam = lyap.exponents().detach().cpu().numpy()
        hks = lyap.hks_proxy().item()

        logs["train_loss"].append(train_loss)
        logs["lambda"].append(lam.copy())
        logs["hks"].append(hks)
        logs["grad_norm"].append(gnorm_acc / max(1, n_batches))

        print(f"[{ep:02d}] T={T:.3g}  loss={train_loss:.4f}  λ1={lam[0]:+.4f}  hKS≈{hks:.4f}  ⟨||∇||⟩={logs['grad_norm'][-1]:.3f}")

    return model, logs
```

```python
temps = [0.0, 1e-4, 5e-4, 1e-3, 5e-3]
results_T = []

for T in temps:
    print("\n=== Running at T =", T, "===")
    m = Net(d_h=16)
    _, logsT = train_with_true_lyap_temp(m, X, y,
                                         lr=1e-2, epochs=20, batch_size=128,
                                         k_lyap=1, T=T, seed=42)
    results_T.append((T, logsT))

# collect final metrics
final_lam1_T = [logs["lambda"][-1][0] for _, logs in results_T]
final_hks_T  = [logs["hks"][-1]          for _, logs in results_T]
final_loss_T = [logs["train_loss"][-1]   for _, logs in results_T]
```

```python
def cosine_anneal(T0, epochs):
    return [0.5*T0*(1+np.cos(np.pi*e/epochs)) for e in range(epochs)]

def train_anneal_T(model, X, y, lr=1e-2, epochs=30, T0=1e-3, **kw):
    # per-epoch temperature schedule
    Ts = cosine_anneal(T0, epochs)
    all_T, all_lam1, all_hks, all_loss = [], [], [], []
    m = model
    # warm start: a couple epochs at T0 to initialize
    for e, T in enumerate(Ts, 1):
        _, logs = train_with_true_lyap_temp(m, X, y, lr=lr, epochs=1, T=T, **kw)
        all_T.append(T)
        all_lam1.append(logs["lambda"][-1][0])
        all_hks.append(logs["hks"][-1])
        all_loss.append(logs["train_loss"][-1])
    return {"T": all_T, "lam1": all_lam1, "hks": all_hks, "loss": all_loss}

logs_ann = train_anneal_T(Net(d_h=16), X, y, lr=1e-2, epochs=30, T0=1e-3,
                          batch_size=128, k_lyap=1, micro=64, device="cpu", seed=123)

# visualize the annealing trajectory
fig, ax = plt.subplots(3,1, figsize=(7,9), sharex=True)
ax[0].plot(logs_ann["T"]); ax[0].set_ylabel("T")
ax[1].plot(logs_ann["lam1"]); ax[1].axhline(0, color='k', lw=1, ls='--'); ax[1].set_ylabel("λ₁")
ax[2].plot(logs_ann["loss"]); ax[2].set_ylabel("Loss"); ax[2].set_xlabel("Epoch")
ax[0].set_title("Cosine-annealed Temperature → Stability and Loss")
plt.tight_layout(); plt.show()
```
